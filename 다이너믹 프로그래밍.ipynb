# 개미 전사

```python
n = int(input())

array = list(map(int, input().split()))
```

```python
best = [0]*len(array)
for i in range(len(array)):
    if i == 0:
        best[i] = array[i]
    elif i == 1:
        if array[i-1] < array[i]:
            best[i] = array[i]
    else:
        if best[i-2] + array[i] > best[i-1]:
            best[i] = best[i-2] + array[i]
        else:
            best[i] = best[i-1]       
        
            
```

# 효율적인 화폐 구성 



```python
N, M = map(int, input().split())
a = []
for i in range(N):
    a += (list(map(int, input())))
```

```python
a
```

```python
dp = [10001]*(M+1)
dp[0] = 0
```

```python
dp
```

```python
for money in a:
    for x in range(1, M+1):
        if (x-money)>=0 and dp[x-money]!=10001:
            dp[x] = min(dp[x], dp[x-money]+1)            
```

```python
dp
```

```python
if dp[-1]!=10001:
    print(dp[-1])
else:
    print(-1)
```

# 금광 



```python
for tc in range(int(input())):
    n,m = map(int, input().split())
    array = list(map(int, input().split()))
    
    dp = []
    index = 0
    for i in range(n):
        dp.append(array[index:index+m])
        index+=m
    for x in range(1,m):
        for y in range(n):
            if y==0:
                dp[y][x] = dp[y][x] + max(dp[y][x-1], dp[y+1][x-1])
            elif y==(n-1):
                dp[y][x] = dp[y][x] + max(dp[y][x-1], dp[y-1][x-1])
            else:
                dp[y][x] = dp[y][x] + max(dp[y][x-1], dp[y-1][x-1], dp[y+1][x-1])
    result = 0 
    for i in range(n):
        result = max(result, dp[i][-1])
    print(result)
```

# 병사 배치하기 

Longest Increasing Subsequence(LIS)

```python
n = int(input())
array = list(map(int, input().split()))
```

```python
array
```

```python
new_dp = [1]* n
```

```python
new_dp
```

```python
new_array = array[::-1]
new_array
```

```python
for i in range(1,n):
    for j in range(0,i):
        if new_array[j] < new_array[i]:
            new_dp[i] = max(new_dp[j]+1, new_dp[i])
```

```python
for i in range(1,n):
    for j in range(i-1,0,-1):
        if new_array[j] < new_array[i]:
            new_dp[i] = max(new_dp[j]+1, new_dp[i])
            break
```

```python
print(n-max(new_dp))
```
